```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/StudentsPerformance.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
```

```python
# Boxplot by gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score by Gender')
plt.show()

# Histogram of scores
df[['math score', 'reading score', 'writing score']].hist(bins=20, figsize=(10, 5))
plt.suptitle("Distribution of Scores")
plt.show()

# Bar plot for test preparation
sns.barplot(x='test preparation course', y='reading score', data=df)
plt.title('Reading Score vs Test Preparation')
plt.show()
```
